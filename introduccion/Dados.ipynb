```python
from matplotlib import pyplot as plt
from collections import Counter
import random
```

```python
generate_random = lambda number_range: ((random.randint(1, 6)) + (random.randint(1, 6)) for _ in range(number_range))
```

```python
histogram_100 = dict(Counter(generate_random(100)))
histogram_1_000 = dict(Counter(generate_random(1_000)))
histogram_10_000 = dict(Counter(generate_random(10_000)))
```

```python
fig, axs = plt.subplots(3, 1, figsize=(15, 15))

def generate_histograms(current_axs: plt.Axes, histogram: dict, title: str):
    current_axs.bar(histogram.keys(), histogram.values())
    current_axs.title.set_text(title)
    current_axs.grid()

generate_histograms(axs[0], histogram_100, 'Histogram 100 iterations')
generate_histograms(axs[1], histogram_1_000, 'Histogram 1_000 iterations')
generate_histograms(axs[2], histogram_10_000, 'Histogram 10_000 iterations')
plt.show()
```
