```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from collections import Counter
import random
```

```python
choose_random = lambda number_range: list((random.randint(1, 6) for _ in range(number_range)))
```

```python
histogram_100 = dict(Counter(choose_random(100)))
histogram_1_000 = dict(Counter(choose_random(1_000)))
histogram_10_000 = dict(Counter(choose_random(10_000)))
```

```python
fig, axs = plt.subplots(3, 1, figsize=(15,15))

axs[0].bar(histogram_100.keys(), histogram_100.values())
axs[0].title.set_text('Histogram 100 iterations')
axs[0].grid()

axs[1].bar(histogram_1_000.keys(), histogram_1_000.values())
axs[1].title.set_text('Histogram 1_000 iterations')
axs[1].grid()

axs[2].bar(histogram_10_000.keys(), histogram_10_000.values())
axs[2].title.set_text('Histogram 10_000 iterations')
axs[2].grid()

plt.show()
```
